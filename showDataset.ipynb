```python
import numpy as np
import pandas as pd
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
```

```python
print("numpy version : {}".format(np.__version__))
print("numpy random seed : {}".format(777))
```

```python
np.random.seed(777)

normal_instance = 500
seen_abnormal_instance = 300
unseen_abnormal_instance = 200

normal_cluster = pd.DataFrame(data={"x":np.random.normal(-7, 2, normal_instance), "y":np.random.normal(7, 2, normal_instance)})
seen_abnormal_cluster = pd.DataFrame(data={"x":np.random.normal(0, 2, seen_abnormal_instance), "y":np.random.normal(-5, 2, seen_abnormal_instance)})
unseen_abnormal_cluster = pd.DataFrame(data={"x":np.random.normal(1, 1, unseen_abnormal_instance), "y":np.random.normal(6, 1, unseen_abnormal_instance)})
```

```python
ax = sns.scatterplot(x="x", y="y", data=normal_cluster)
ax = sns.scatterplot(x="x", y="y", data=seen_abnormal_cluster)
ax = sns.scatterplot(x="x", y="y", data=unseen_abnormal_cluster)
```
